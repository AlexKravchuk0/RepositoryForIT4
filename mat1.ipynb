```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("StudentsPerformance.csv")
```

```python
data.head()
```

```python
newDF = data.rename(columns = {"parental level of education" : "parentallevelofeducation" ,
                              "test preparation course" : "testpreparationcourse",
                              "math score" : "mathscore",
                              "reading score" : "readingscore",
                              "writing score" : "writingscore"})

# Перейминування всіх колонок де є пробіли
```

```python
newDF.mathscore.hist() # Простий графік з pandas 
```

```python
# Скатер плот :
newDF.plot.scatter(x="mathscore",y="readingscore")
```

```python
import seaborn as sb
```

```python
# Хороша бібліотека для графіків seaborn. Вже з кореляційною прямою 
sb.lmplot(x="mathscore",y="readingscore", data = newDF) 
```

```python
# Хороша бібліотека для графіків seaborn. Вже з кореляційною прямою для двох гендерів
sb.lmplot(x="mathscore",y="readingscore", hue = 'gender', data = newDF) 
```
